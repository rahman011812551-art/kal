# Sistem Persamaan Linier Tiga Variabel (SPLTV)
## Metode Eliminasi Gauss

### Soal

Selesaikan sistem berikut:

x + y + z = 6  
2x + y + z = 9  
x + 2y + 3z = 14  

---

## Langkah 1: Matriks Augmented

\[
\begin{bmatrix}
1 & 1 & 1 & | & 6 \\
2 & 1 & 1 & | & 9 \\
1 & 2 & 3 & | & 14
\end{bmatrix}
\]

---

## Langkah 2: Eliminasi

Hilangkan elemen di bawah pivot pertama:

R2 ŌåÆ R2 ŌłÆ 2R1  
R3 ŌåÆ R3 ŌłÆ R1  

\[
\begin{bmatrix}
1 & 1 & 1 & | & 6 \\
0 & -1 & -1 & | & -3 \\
0 & 1 & 2 & | & 8
\end{bmatrix}
\]

Hilangkan elemen di bawah pivot kedua:

R3 ŌåÆ R3 + R2  

\[
\begin{bmatrix}
1 & 1 & 1 & | & 6 \\
0 & -1 & -1 & | & -3 \\
0 & 0 & 1 & | & 5
\end{bmatrix}
\]

---

## Langkah 3: Substitusi Mundur

Dari baris ketiga:

z = 5  

Baris kedua:

ŌłÆy ŌłÆ z = ŌłÆ3  
ŌłÆy ŌłÆ 5 = ŌłÆ3  
y = ŌłÆ2  

Baris pertama:

x + y + z = 6  
x ŌłÆ 2 + 5 = 6  
x = 3  

---

## Hasil SPLTV

x = 3  
y = ŌłÆ2  
z = 5

# Sistem Persamaan Linier Empat Variabel (SPL4V)
## Metode Eliminasi Gauss

### Soal

x + y + z + w = 10  
2x + y + z + w = 12  
x + 2y + z + w = 13  
x + y + 2z + 3w = 20  

---

## Langkah 1: Matriks Augmented

\[
\begin{bmatrix}
1 & 1 & 1 & 1 & | & 10 \\
2 & 1 & 1 & 1 & | & 12 \\
1 & 2 & 1 & 1 & | & 13 \\
1 & 1 & 2 & 3 & | & 20
\end{bmatrix}
\]

---

## Langkah 2: Eliminasi Kolom Pertama

R2 ŌåÆ R2 ŌłÆ 2R1  
R3 ŌåÆ R3 ŌłÆ R1  
R4 ŌåÆ R4 ŌłÆ R1  

\[
\begin{bmatrix}
1 & 1 & 1 & 1 & | & 10 \\
0 & -1 & -1 & -1 & | & -8 \\
0 & 1 & 0 & 0 & | & 3 \\
0 & 0 & 1 & 2 & | & 10
\end{bmatrix}
\]

---

## Langkah 3: Eliminasi Kolom Kedua

R3 ŌåÆ R3 + R2  

\[
\begin{bmatrix}
1 & 1 & 1 & 1 & | & 10 \\
0 & -1 & -1 & -1 & | & -8 \\
0 & 0 & -1 & -1 & | & -5 \\
0 & 0 & 1 & 2 & | & 10
\end{bmatrix}
\]

---

## Langkah 4: Eliminasi Kolom Ketiga

R4 ŌåÆ R4 + R3  

\[
\begin{bmatrix}
1 & 1 & 1 & 1 & | & 10 \\
0 & -1 & -1 & -1 & | & -8 \\
0 & 0 & -1 & -1 & | & -5 \\
0 & 0 & 0 & 1 & | & 5
\end{bmatrix}
\]

---

## Langkah 5: Substitusi Mundur

Dari baris keempat:

w = 5  

Baris ketiga:

ŌłÆz ŌłÆ w = ŌłÆ5  
ŌłÆz ŌłÆ 5 = ŌłÆ5  
z = 0  

Baris kedua:

ŌłÆy ŌłÆ z ŌłÆ w = ŌłÆ8  
ŌłÆy ŌłÆ 0 ŌłÆ 5 = ŌłÆ8  
y = 3  

Baris pertama:

x + y + z + w = 10  
x + 3 + 0 + 5 = 10  
x = 2  

---

## Hasil SPL4V

x = 2  
y = 3  
z = 0  
w = 5
